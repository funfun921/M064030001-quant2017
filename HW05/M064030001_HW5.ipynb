```python
import pandas_datareader.data as web
from datetime import datetime
import requests 
```

```python
def stock(code,date):
    url = "http://finance.google.com/finance/historical"
    params = {}
    params['q'] = code
    params['startdate'] = date
    params['output'] = 'csv'
    r = requests.get(url, params=params)
    print(r.text)
```

```python
stock('AAPL','2016-01-01')
```
